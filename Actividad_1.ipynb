```python
import pandas as pd
import numpy as np
import random

# Número de registros
n = 500

# Listas de valores categóricos
generos = ['Masculino', 'Femenino']
ciudades = ['Barranquilla', 'Bogotá', 'Medellín', 'Cali', 'Cartagena']
nivel_socio = ['Bajo', 'Medio', 'Alto']
beca = ['Si', 'No']

# Crear datos
data = {
    'edad': np.random.randint(16, 30, n),
    'genero': np.random.choice(generos, n),
    'ciudad': np.random.choice(ciudades, n),
    'promedio_colegio': np.round(np.random.uniform(2.0, 5.0, n), 2),
    'puntaje_examen': np.random.randint(100, 500, n),
    'nota_semestre': np.round(np.random.uniform(2.0, 5.0, n), 2),
    'nivel_socioeconomico': np.random.choice(nivel_socio, n),
    'beca': np.random.choice(beca, n)
}

df = pd.DataFrame(data)

# Crear variable objetivo (deserción)
df['desercion'] = np.where(
    (df['nota_semestre'] < 3.0) |
    (df['promedio_colegio'] < 3.0) |
    (df['nivel_socioeconomico'] == 'Bajo'),
    'Si', 'No'
)

# 🔴 Introducir valores nulos (5%)
for col in df.columns:
    df.loc[df.sample(frac=0.05).index, col] = np.nan

# 🔴 Introducir outliers
df.loc[df.sample(10).index, 'nota_semestre'] = np.random.uniform(0.0, 1.5, 10)
df.loc[df.sample(10).index, 'puntaje_examen'] = np.random.randint(0, 50, 10)

# Guardar dataset
df.to_csv('dataset_desercion.csv', index=False)

print("Dataset generado correctamente")
df.head()
```
